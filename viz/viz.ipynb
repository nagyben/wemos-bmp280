```python
from google.cloud import firestore
import pandas

db = firestore.Client(project="nagyben")
```

```python
docs = db.collection(u"weather-test").stream()

df = pandas.DataFrame(next(docs).to_dict()["data"])

for doc in docs:
    df = pandas.concat([df, pandas.DataFrame(doc.to_dict()["data"])])

df
```

```python
df["timestamp"] = pandas.to_datetime(df["timestamp"])
```

```python
df = df.loc[df["git_rev"] == df["git_rev"].iloc[-1]]
df
```

```python
# with 100kOhm resistor
# 913 == 3.62V
# 879 == 3.48V

# with 129kOhm resistor
# 946 == 4.11V
# 942 == 4.09V
# 823 == 3.56V
x1 = 946
y1 = 4.11
x2 = 823
y2 = 3.56
m = (y1-y2)/(x1-x2)
# y = mx + c
# c = y - mx
c = y1 - (m * x1)

def get_vcc(adc: int) -> float:
    return m * adc + c
```

```python
# max voltage that can be measured
print(f"Max measurable voltage: {1024 * m + c:.3f}V")
```

```python
cur_voltage = get_vcc(df["Vcc"].iloc[-1])
print(f"Current voltage: {cur_voltage:.3f}V")
```

```python
import plotly.express as px
import plotly.graph_objects as go
import datetime
```

```python
template = dict(
    layout=go.Layout(
        font=dict(family="Maven Pro")
    )
)
```

```python
# forecasting
WINDOW = 12 * 24
import scipy.stats

linregress = scipy.stats.linregress(x = df["timestamp"].astype('int64')[-WINDOW:], y=get_vcc(df["Vcc"])[-WINDOW:])
linregress
WINDOW
```

```python

forecast_t = pandas.date_range(df["timestamp"].astype('int64').iloc[-WINDOW], end=datetime.date.today() + datetime.timedelta(days=10), freq="5T")
forecast_vcc = linregress.slope * forecast_t.astype('int64') + linregress.intercept
forecast_vcc
```

```python
# y = mx + c
# x = (y - c) / m
forecast_3_6v_time = pandas.Timestamp((3.6 - linregress.intercept) / linregress.slope, unit="ns")
forecast_3_6v_time
```

```python

from plotly.subplots import make_subplots

fig = go.Figure(layout=go.Layout(
    yaxis=dict(range=[3.2, 4.2], title="Vcc"),
    yaxis2=dict(overlaying="y", side="right", title="discharge rate"),
))
fig.add_trace(go.Scatter(
    x=df["timestamp"],
    y=get_vcc(df["Vcc"]).rolling(window=12).mean(),
    name="Vcc",
    mode="markers",
    marker=dict(size=2)
))
fig.add_trace(go.Scatter(
    x=[df["timestamp"].min(), forecast_t.max()] + [forecast_t.max(), df["timestamp"].min()],
    y=[3.6, 3.6],
    name="nominal voltage",
    mode="lines",
    line=dict(dash="dash", color="rgba(0,0,0,1)", width=1),
))
fig.add_trace(go.Scatter(
    x=[df["timestamp"].min(), forecast_t.max()] + [forecast_t.max(), df["timestamp"].min()],
    y=[3.3, 3.3, 0, 0],
    name="3.3v limit",
    mode="lines",
    fill="toself",
    fillcolor="rgba(255,0,0,0.2)",
    line=dict(width=0),
))
fig.add_trace(go.Scatter(
    x=forecast_t,
    y=forecast_vcc,
    name="forecast",
    mode="lines",
    line=dict(width=1),
))
fig.add_annotation(
    text="<b>3.6v nominal</b>",
    x=df["timestamp"].iloc[int(len(df)/2)],
    y=3.63,
    showarrow=False,
    font=dict(
        color="rgba(0,0,0,1)",
        family="Maven Pro"
    )
)
fig.add_annotation(
    text="<b>3.3v limit</b>",
    x=df["timestamp"].iloc[int(len(df)/2)],
    y=3.33,
    showarrow=False,
    font=dict(
        color="rgba(255,0,0,1)",
        family="Maven Pro"
    )
)
fig.add_annotation(
    text=f"{forecast_3_6v_time.strftime('%d-%b @ %H:%M ')}",
    x=forecast_3_6v_time,
    y=3.6,
    font=dict(
        color="rgba(255,0,0,1)",
        family="Maven Pro"
    )
)
# fig.add_annotation(
#     text=f"Current voltage: {df['timestamp'].iloc[-1].astype(float)}",
#     x=df["timestamp"].iloc[-1],
#     y=df["Vcc"].iloc[-1],
#     font=dict(
#         color="rgba(255,0,0,1)",
#         family="Maven Pro"
#     )
# )
fig.update_layout(template=template)
fig.show()
```

```python

traces = [
    go.Scatter(
    x=df["timestamp"],
    y=df["temp_C"],
    name="temp_C",
    hovertemplate='<b>%{x}</b><br>%{y:.1f}',
),
go.Scatter(
    x=df["timestamp"],
    y=df["pressure_Pa"]/100,
    name="pressure_mbar",
    yaxis="y2",
    hovertemplate='<b>%{x}</b><br>%{y:.0f}',
),
go.Scatter(
    x=df["timestamp"],
    y=df["humidity_%"],
    name="humidity_%",
    yaxis="y3",
    hovertemplate='<b>%{x}</b><br>%{y:.1f}',
),
go.Scatter(
    x=df["timestamp"],
    y=df["Vcc"],
    name="Vcc",
    yaxis="y4",
    hovertemplate='<b>%{x}</b><br>%{y:.0f}',
),
go.Scatter(
    x=df["timestamp"],
    y=df["postSensorRead_ms"],
    name="postSensorRead_ms",
    yaxis="y5",
    hovertemplate='<b>%{x}</b><br>%{y:.0f}',
),
go.Scatter(
    x=df["timestamp"],
    y=df["postConnectTime_ms"].rolling(window=12*24).mean(),
    name="postConnectTime_ms",
    yaxis="y6",
    hovertemplate='<b>%{x}</b><br>%{y:.0f}',
),
]

fig = make_subplots(rows=len(traces), cols=1, shared_xaxes=True, vertical_spacing=0.01, start_cell="bottom-left", )

for trace in traces:
    fig.add_trace(trace)

fig.update_layout(template="none", hovermode="x")

fig.show()
```
